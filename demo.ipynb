```python
import requests
from bs4 import BeautifulSoup
import yfinance as yf
import datetime
import nltk
from transformers import pipeline
import matplotlib.pyplot as plt

nltk.download('punkt')

# Load sentiment analyzer (pretrained)
sentiment_pipeline = pipeline("sentiment-analysis")

def fetch_latest_news(stock_name):
    print(f"\n Fetching latest news for: {stock_name}")
    url = f"https://www.moneycontrol.com/news/tags/{stock_name.replace(' ', '-')}.html"
    headers = {'User-Agent': 'Mozilla/5.0'}
    
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.content, 'html.parser')
    
    headlines = soup.select(".clearfix h2")[:5]  # Get top 5 headlines
    news = [tag.get_text(strip=True) for tag in headlines if tag.get_text(strip=True)]
    return news

def analyze_sentiment(news_list):
    print("\n Analyzing sentiment of news...")
    sentiments = sentiment_pipeline(news_list)
    for n, s in zip(news_list, sentiments):
        print(f"{n}  {s['label']} ({s['score']:.2f})")
    return sentiments

def get_stock_price_movement(ticker):
    today = datetime.date.today()
    start = today - datetime.timedelta(days=5)
    end = today + datetime.timedelta(days=1)

    stock = yf.Ticker(ticker)
    hist = stock.history(start=start, end=end)

    if hist.empty:
        print(" No stock data found!")
        return None


    # Calculate percentage change
    close_prices = hist['Close']
    percent_change = ((close_prices[-1] - close_prices[0]) / close_prices[0]) * 100

    print(f"\n Stock price movement over last few days: {percent_change:.2f}%")
    return percent_change

def predict_news_impact(sentiments, price_change):
    # Count sentiment labels
    positive = sum(1 for s in sentiments if s['label'] == 'POSITIVE')
    negative = sum(1 for s in sentiments if s['label'] == 'NEGATIVE')

    print(f"\n Positive News: {positive}, Negative News: {negative}")
    
    if price_change > 0 and positive >= negative:
        impact = "Likely Positive Impact"
    elif price_change < 0 and negative > positive:
        impact = "Likely Negative Impact"
    else:
        impact = "Neutral or Mixed Impact"
    
    print(f"\n Predicted news impact on stock: {impact}")
    return impact

# --------------------------
#  MAIN PROGRAM
# --------------------------
if __name__ == "__main__":
    stock_name = "Reliance"
    ticker_symbol = "RELIANCE.NS"

    news = fetch_latest_news(stock_name)
    if not news:
        print("No news found.")
        exit()

    sentiments = analyze_sentiment(news)
    price_change = get_stock_price_movement(ticker_symbol)

    if price_change is not None:
        predict_news_impact(sentiments, price_change)
```
